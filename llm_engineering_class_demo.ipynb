# LLM Engineering Class 1

This notebook demonstrates prompt engineering and a basic ChromaDB RAG pipeline.

```python
from llm_utils import load_llm, ask_llm
from rag_chromadb_demo import build_vector_store, retrieve_top_k, build_rag_prompt
```

## Prompt Engineering Example

```python
generator = load_llm()

prompt = """
You are an expert Python teacher.
Explain what a Python dictionary is for a beginner.
Include one simple example.
""".strip()

answer = ask_llm(generator, prompt)
print(answer)
```

## ChromaDB + RAG Example

```python
collection = build_vector_store()
question = "What is ChromaDB used for?"
retrieved_docs = retrieve_top_k(collection, question, top_k=2)

print("Retrieved Documents:\n")
for i, doc in enumerate(retrieved_docs, start=1):
    print(f"{i}. {doc}\n")
```

```python
rag_prompt = build_rag_prompt(question, retrieved_docs)
rag_answer = ask_llm(generator, rag_prompt, max_new_tokens=160)

print(rag_answer)
```
